## 資料預處理
### 1.球種名稱更改


```python
name_mapping = {
    'SL':'Slider',
    'SI':'Sinker', 
    'FF':'Fastball', 
    'CH':'ChangeUp', 
    'CU':'Curveball', 
    'FO':'Splitter', 
    'FC':'Cutter',
    'KN':'Knuckle curve',
    'FS':'Splitter'
    }

# 新增一列"球員"，根据"Pitcher"列的值匹配中文名
data ['球種'] = data ['pType'].map(name_mapping)
```

### 2.球隊ID 轉換

```python
name_mapping = {
    '65630ee55559b6d7a6d5d200':'兄弟',
    '65630ee95559b6d7a6d5d201':'味全', 
    '65630eed5559b6d7a6d5d202':'樂天', 
    '65630ef25559b6d7a6d5d203':'統一',
    '65630ef65559b6d7a6d5d204':'富邦',
    '6569a9ddaf62a48d3f94b422':'台鋼',
    }

# 新增一列"球員"，根据"Pitcher"列的值匹配中文名
#data ['球隊'] = data ['bTeamId'].map(name_mapping)
```

### 3.每打席每顆球編號

```python
data['PitchofPA'] = data.groupby(['bName', 'pName', 'date', 'gameId', 'paRound', 'paOrder']).cumcount() + 1
```

### 4.標記下一顆球

```python
data.sort_values(by=['bName', 'pName', 'date', 'gameId', 'paRound', 'paOrder', 'PitchofPA'], inplace=True)

# 新增 "Next_Pitch" 欄位，將下一筆資料的 "球種" 填入
data['Next_Pitch'] = data['球種'].shift(-1)
last_pitch_indices = data.groupby(['bName', 'pName', 'date', 'gameId', 'paRound', 'paOrder']).tail(1).index
data.loc[last_pitch_indices, 'Next_Pitch'] = None
```

### 5.標記壘包狀態

```python
def base_state(row):
    state = 0
    if not pd.isna(row['firstId']):
        state += 1
    if not pd.isna(row['secondId']):
        state += 2
    if not pd.isna(row['thirdId']):
        state += 4
    return state

data['壘包狀態'] = data.apply(base_state, axis=1)
```

### 6.分差(客場-主場)

```python
data['分差']=data['awayScores']-data['homeScores']
```

### 7.重新篩選欄位

```python
col2 = ['date','pName','bName','cName','bHand','paRound','paOrder','壘包狀態','velo','call','isStrike','球種','分差','inning','side','Next_Pitch']
```

```python
df = data[col2]
```

```python
df = df[df['Next_Pitch'].notna()]
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
from sklearn.preprocessing import LabelEncoder
# 转换分类特征为数值编码

label_encoders = {}
for column in ['pName', 'bName', 'cName', 'bHand', 'call', '球種', 'side']:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le

# 对目标变量进行编码
le_target = LabelEncoder()
df['Next_Pitch'] = le_target.fit_transform(df['Next_Pitch'])
```

```python
df = df.fillna('')
```

## 預測環節

### 一.Train,Test切分

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score
```

```python
X = df.drop(columns=['isStrike','Next_Pitch'])
y = df['Next_Pitch']

X = X.apply(pd.to_numeric, errors='coerce')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=40)
```

```python
print(X_train.dtypes)
```

### 二.預測
### 1.RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(max_depth = 30, n_estimators = 200, random_state=40)
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
y_pred_decoded = le_target.inverse_transform(y_pred)
y_test_decoded = le_target.inverse_transform(y_test)
```

```python
print('Predictions:', y_pred_decoded)
print('Actual:', y_test_decoded)
```

### 2.Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth = 30, random_state=40)
model.fit(X_train, y_train)
```

```python
y_pred_tree = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred_tree)
print(f'Accuracy: {accuracy:.2f}')
```

```python
y_pred_decoded = le_target.inverse_transform(y_pred_tree)
y_test_decoded = le_target.inverse_transform(y_test)
```

```python
print('Predictions:', y_pred_decoded)
print('Actual:', y_test_decoded)
```
