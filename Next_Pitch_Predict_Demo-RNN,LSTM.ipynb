## 資料預處理
### 1.球種名稱更改


```python
name_mapping = {
    'SL':'Slider',
    'SI':'Sinker', 
    'FF':'Fastball', 
    'CH':'ChangeUp', 
    'CU':'Curveball', 
    'FO':'Splitter', 
    'FC':'Cutter',
    'KN':'Knuckle curve',
    'FS':'Splitter'
    }

# 新增一列"球員"，根据"Pitcher"列的值匹配中文名
data ['球種'] = data ['pType'].map(name_mapping)
```

### 2.球隊ID 轉換

```python
name_mapping = {
    '65630ee55559b6d7a6d5d200':'兄弟',
    '65630ee95559b6d7a6d5d201':'味全', 
    '65630eed5559b6d7a6d5d202':'樂天', 
    '65630ef25559b6d7a6d5d203':'統一',
    '65630ef65559b6d7a6d5d204':'富邦',
    '6569a9ddaf62a48d3f94b422':'台鋼',
    }

# 新增一列"球員"，根据"Pitcher"列的值匹配中文名
#data ['球隊'] = data ['bTeamId'].map(name_mapping)
```

### 3.每打席每顆球編號

```python
data['PitchofPA'] = data.groupby(['bName', 'pName', 'date', 'gameId', 'paRound', 'paOrder']).cumcount() + 1
```

### 4.標記下一顆球

```python
data.sort_values(by=['bName', 'pName', 'date', 'gameId', 'paRound', 'paOrder', 'PitchofPA'], inplace=True)

# 新增 "Next_Pitch" 欄位，將下一筆資料的 "球種" 填入
data['Next_Pitch'] = data['球種'].shift(-1)
last_pitch_indices = data.groupby(['bName', 'pName', 'date', 'gameId', 'paRound', 'paOrder']).tail(1).index
data.loc[last_pitch_indices, 'Next_Pitch'] = None
```

### 5.標記壘包狀態

```python
def base_state(row):
    state = 0
    if not pd.isna(row['firstId']):
        state += 1
    if not pd.isna(row['secondId']):
        state += 2
    if not pd.isna(row['thirdId']):
        state += 4
    return state

data['壘包狀態'] = data.apply(base_state, axis=1)
```

### 6.分差(客場-主場)

```python
data['分差']=data['awayScores']-data['homeScores']
```

## 預測環節

### 一.Train,Test切分

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Dropout,LSTM
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
cat_cols = ['pName', 'bName', 'cName', 'bHand', 'call', '球種', 'side', 'Next_Pitch','isStrike']

label_encoders = {}
for col in cat_cols:
    label_encoders[col] = LabelEncoder()
    df[col] = label_encoders[col].fit_transform(df[col].astype(str))
```

```python
X = df.drop(columns=['Next_Pitch'])
y = df['Next_Pitch']

X = X.apply(pd.to_numeric, errors='coerce')
```

```python
X = X.values
y = y.values

#X = X.apply(pd.to_numeric, errors='coerce')
# 分割数据集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=40)
```

### 1.RNN

```python
# Create a time series generator for training
n_input = 3  # Length of the input sequences
generator_train = TimeseriesGenerator(X_train, y_train, length=n_input, batch_size=32)

# Create a time series generator for validation
generator_val = TimeseriesGenerator(X_test, y_test, length=n_input, batch_size=32)
```

```python
model = Sequential()
model.add(SimpleRNN(50, activation='relu', input_shape=(n_input, X_train.shape[1])))
model.add(Dropout(0.2))
model.add(Dense(13, activation='softmax')) 
```

```python
# 編譯模型
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 訓練模型
model.fit(generator_train, epochs=20, validation_data=generator_val)
```

```python
predictions = model.predict(generator_val)

# Evaluate the model
accuracy = model.evaluate(generator_val)[1]
print(f'Accuracy: {accuracy * 100:.2f}%')
```

### 2.LSTM

```python
# Create a time series generator for training
n_input = 5  # Length of the input sequences
generator_train = TimeseriesGenerator(X_train, y_train, length=n_input, batch_size=1)

# Create a time series generator for validation
generator_val = TimeseriesGenerator(X_test, y_test, length=n_input, batch_size=1)
```

```python
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(n_input, X_train.shape[1])))
model.add(Dropout(0.2))
model.add(Dense(13, activation='softmax')) 
```

```python
# 編譯模型
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 訓練模型
model.fit(generator_train, epochs=20, validation_data=generator_val)
```

```python
predictions = model.predict(generator_val)

# Evaluate the model
accuracy = model.evaluate(generator_val)[1]
print(f'Accuracy: {accuracy * 100:.2f}%')
```
